```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

### 1. Countplot

```python
sns.countplot(x = df['Survived'], data = df,stat = 'percent',hue='Survived')
```

```python
df['Survived'].value_counts().plot(kind = 'bar')
```

```python
sns.countplot(x = df['Cabin'], data = df)
```

### PiChart

```python
df['Sex'].value_counts().plot(kind = 'pie', autopct = '%.2f')
```
