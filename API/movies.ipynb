```python
import requests
import pandas as pd
import time

all_data = []

for i in range(1, 508):
    url = f"https://api.themoviedb.org/3/movie/top_rated?api_key=fbf85508700a25045bc8add07e80d431&language=en-US&page={i}"
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Raise exception for HTTP errors
        results = response.json().get('results', [])
        temp_df = pd.DataFrame(results)[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        all_data.append(temp_df)
        print(f"Page {i} fetched successfully.")
        time.sleep(0.5)  # Avoid hitting API rate limits
    except requests.exceptions.Timeout:
        print(f"Timeout occurred on page {i}. Retrying...")
        time.sleep(5)
        continue
    except requests.exceptions.RequestException as e:
        print(f"Request failed on page {i}: {e}")
        continue

# Combine all the DataFrames
df = pd.concat(all_data, ignore_index=True)
```
